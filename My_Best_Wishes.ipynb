```python
# My wishes for this new year 2020
import csv
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('My_wishes.csv')
df
```

```python
wishes = df['Happy New Year from Veronique Poudima. 2020']
wishes
```

```python
wishList = []
for i in wishes:
    wishList.append(i.split())
```

```python
wish = []
for i in wishList:
    for a in i:
        wish .append(a)
```

```python
from wordcloud import WordCloud
wordcould_dict= Counter(wish)
wordcloud = WordCloud(width = 1000, height = 500).generate_from_frequencies(wordcould_dict)

plt.figure(figsize=(20,10))
plt.figtext(.5,.9,"Best Wishes for 2020", fontsize=60, ha='center')
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
